```python
import pandas as pd
import numpy as np
import requests
```

```python
url = "https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=orcl&outputsize=full&apikey=SVRXBBY37Z6W0BJZ"
```

```python
r = requests.get(url).json()
```

```python
r
```

```python
df = pd.DataFrame(r['Time Series (Daily)'])
df = df.T
```

```python
df.tail()
```

```python
date = []
for row in df.index:
    date.append(row)
```

```python
df['date']=date
```

```python
df.set_index(np.arange(0,len(df)), inplace = True)
```

```python
df = df.iloc[31:4562]
```

```python
df.set_index(np.arange(0,len(df)), inplace = True)
```

```python
df.head()
```

```python
sym = r['Meta Data']['2. Symbol']
```

```python
symb = [sym]*len(df)
```

```python
df['symbol'] = symb
```

```python
year = []
month = []
quarter = []
popen = []
phigh = []
plow = []
pclose = []
pvolume = []
for i in range(len(df)):
    year.append(df.iloc[i]['date'].split("-")[0])
    month.append(df.iloc[i]['date'].split("-")[1])
    if month[i] == '12' or month[i] == '11' or month[i] == '10':
        quarter.append(4)
    elif month[i] == '09'or month[i] == '08' or month[i] == '07':
        quarter.append(3)
    elif month[i] == '06' or month[i] == '05' or month[i] == '04':
        quarter.append(2)
    else:
        quarter.append(1)
    popen.append(float(df.iloc[i]['1. open']))
    phigh.append(float(df.iloc[i]['2. high']))
    plow.append(float(df.iloc[i]['3. low']))
    pclose.append(float(df.iloc[i]['4. close']))
    pvolume.append(float(df.iloc[i]['5. volume']))
    
```

```python
df['year'] = year
```

```python
df['quarter'] = quarter
```

```python
del df['1. open']
del df['2. high']
del df['3. low']
del df['4. close']
del df['5. volume']
```

```python
df['open'] = popen
df['high'] = phigh
df['low'] = plow
df['close'] = pclose
df['volume'] = pvolume
```

```python
df
```

```python
uu = np.where(df['close'] - df['open'] > 0, 1, 0)
df["pup"] = uu
```

```python
vv = []
for i in range(len(df)-1):
    vv.append(df.iloc[i]['volume'] - df.iloc[i+1]['volume'])
    
vv.append(1)
```

```python
ww = []
for j in range(len(vv)):
    if vv[j] > 0:
        vv[j] = 1
    else:
        vv[j] = 0
    if uu[j] == 1 and vv[j] == 1:
        ww.append(1)
    else:
        ww.append(0)
```

```python
df["vup"] = vv
```

```python
df["bothup"] = ww
```
