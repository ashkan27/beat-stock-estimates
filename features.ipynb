```python
import pandas as pd
import numpy as np
```

```python
symbol = ['msft', 'aapl', 'amzn', 'adbe', 'qcom', 'orcl', 'acn', 'ibm', 'xom', 'intu',
         'pg', 'blk', 'ups', 'hd', 'mcd', 't', 'a', 'wmt', 'dis', 'tgt',
         'adi', 'jnj', 'mrk', 'mdt', 'syk', 'cl', 'el', 'ko', 'mmm', 'nke',
         'utx', 'lmt', 'axp', 'spg', 'mmc', 'all', 'k', 'low', 'gild', 'yum',
         'pld', 'stt', 'ip', 'nee', 'shw', 'ea', 'abt', 'unh', 'vrsn', 'lly']

symbol.sort(reverse=True)
```

```python
raws_csv = []

for s in symbol:
    # Loop through the Features_Data folder, add ticker column to each csv with each read then append
    raw_csv = pd.read_csv(f'raw/{s}_raw.csv')
    raws_csv.append(raw_csv)
    
raw_df = pd.concat(raws_csv)
```

```python
raw_df.head()
```

```python
len(raw_df)
```

```python
raw_df.to_csv('raw_df.csv', index=False)
```

```python
grouped = raw_df.groupby(['symbol','year', 'quarter'])[['pup']]
sumupday = grouped.sum()['pup']
countupday = grouped.count()['pup']
pup = sumupday/countupday
len(pup)

feature1_pup = []
for i in range(len(pup)):
    feature1_pup.append(pup[-(i+1)])
        
```

```python
feature1_pup[-1]
```

```python
grouped = raw_df.groupby(['symbol','year', 'quarter'])[['vup']]
sumvup = grouped.sum()['vup']
countvup = grouped.count()['vup']
vup = sumvup/countvup
len(vup)

feature2_vup = []
for i in range(len(vup)):
    feature2_vup.append(vup[-(i+1)])
```

```python
feature2_vup[-1]
```

```python
grouped = raw_df.groupby(['symbol','year', 'quarter'])[['bothup']]
sumboth = grouped.sum()['bothup']
countboth = grouped.count()['bothup']
bothup = sumboth/countboth
len(bothup)

feature3_bothup = []
for i in range(len(bothup)):
    feature3_bothup.append(bothup[-(i+1)])
```

```python
feature3_bothup[-5]
```

```python
grouped = raw_df.groupby(['symbol','year', 'quarter'])[['close']]
aveclose = grouped.mean()['close']
len(aveclose)

feature4_aveclose = []
for i in range(len(aveclose)):
    feature4_aveclose.append(aveclose[-(i+1)])
```

```python
feature4_aveclose
```

```python
features_df = pd.read_csv('estimates_df.csv')
```

```python
len(features_df)
```

```python
features_df['f1_pup'] = feature1_pup
features_df['f2_vup'] = feature2_vup
features_df['f3_bothup'] = feature3_bothup
features_df['f4_average_price'] = feature4_aveclose
```

```python
features_df.to_csv('features_df.csv', index=False)
```
