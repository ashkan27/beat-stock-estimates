```python
predictions_csv = []

for s in symbol:
    # Loop through the Features_Data folder, add ticker column to each csv with each read then append
    prediction_csv = pd.read_csv(f'prediction/{s}_prediction.csv')
    predictions_csv.append(prediction_csv)
    
prediction_df = pd.concat(predictions_csv)
```

```python
prediction_df.head()
```

```python
grouped = prediction_df.groupby(['symbol','year', 'quarter'])[['pup']]
sumupday = grouped.sum()['pup']
countupday = grouped.count()['pup']
pup = sumupday/countupday
len(pup)

feature1_pup = []
for i in range(len(pup)):
    feature1_pup.append(pup[(i)])
```

```python
grouped = prediction_df.groupby(['symbol','year', 'quarter'])[['vup']]
sumvup = grouped.sum()['vup']
countvup = grouped.count()['vup']
vup = sumvup/countvup
len(vup)

feature2_vup = []
for i in range(len(vup)):
    feature2_vup.append(vup[(i)])
```

```python
grouped = prediction_df.groupby(['symbol','year', 'quarter'])[['bothup']]
sumboth = grouped.sum()['bothup']
countboth = grouped.count()['bothup']
bothup = sumboth/countboth
len(bothup)

feature3_bothup = []
for i in range(len(bothup)):
    feature3_bothup.append(bothup[(i)])
```

```python
grouped = prediction_df.groupby(['symbol','year', 'quarter'])[['close']]
aveclose = grouped.mean()['close']
len(aveclose)

feature4_aveclose = []
for i in range(len(aveclose)):
    feature4_aveclose.append(aveclose[(i)])
```

```python
prediction_data = pd.DataFrame({'symbol': symbol, 'estimate' : [2.81, 6.43, 0.98, 1.02, 1.33]})
```

```python
prediction_data
```

```python
prediction_data['f1_pup'] = feature1_pup
prediction_data['f2_vup'] = feature2_vup
prediction_data['f3_bothup'] = feature3_bothup
prediction_data['f4_average_price'] = feature4_aveclose
```

```python
prediction_data
```

## Using Neural Network Model

```python
from tensorflow.keras.models import load_model
model = load_model("stock1.h5")
```

```python
X = prediction_data[['f1_pup', 'f2_vup', 'f3_bothup', 'f4_average_price', 'estimate']]
```

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler().fit(X)
X_scaled = scaler.transform(X)
```

```python
predictions_nn = model.predict_classes(X_scaled)
```

```python
prediction_data['Prediction Beat? (NN)'] = predictions_nn
```
