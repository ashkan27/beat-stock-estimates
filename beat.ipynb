```python
import pandas as pd
import numpy as np
```

```python
symbol = ['msft', 'aapl', 'amzn', 'adbe', 'qcom', 'orcl', 'acn', 'ibm', 'xom', 'intu',
         'pg', 'blk', 'ups', 'hd', 'mcd', 't', 'a', 'wmt', 'dis', 'tgt',
         'adi', 'jnj', 'mrk', 'mdt', 'syk', 'cl', 'el', 'ko', 'mmm', 'nke',
         'utx', 'lmt', 'axp', 'spg', 'mmc', 'all', 'k', 'low', 'gild', 'yum',
         'pld', 'stt', 'ip', 'nee', 'shw', 'ea', 'abt', 'unh', 'vrsn', 'lly']

symbol.sort(reverse=True)
```

```python
estimates_csv = []

for s in symbol:
    # Loop through the Features_Data folder, add ticker column to each csv with each read then append
    estimate_csv = pd.read_csv(f'stocks/{s}_estimate.csv')
    estimates_csv.append(estimate_csv)
    
estimates_df = pd.concat(estimates_csv)
```

```python
estimates_df
```

```python
year = []
for i in range(len(estimates_df)):
    year.append(estimates_df.iloc[i]['period'].split("-")[0])
```

```python
estimates_df['year'] = year
```

```python
month = []
quarter = []
for i in range(len(estimates_df)):
    month.append(estimates_df.iloc[i]['period'].split("-")[1])
    if month[i] == '12':
        quarter.append(4)
    elif month[i] == '09':
        quarter.append(3)
    elif month[i] == '06':
        quarter.append(2)
    else:
        quarter.append(1)
        
    
```

```python
estimates_df['quarter'] = quarter
```

```python
estimates_df['beat'] = np.where(estimates_df['actual'] - estimates_df['estimate'] >= 0, 1, 0)
```

```python
estimates_df.to_csv('estimates_df.csv', index=False)
```
